## COUNTRY DATA

```python
df_country=pd.read_csv('countries.csv')
```

```python
df_country
```

```python
us=df_country[df_country.country=='United States']
```

```python
us
```

```python
china=df_country[df_country.country=='China']
```

```python
china
```

```python
plt.plot(us.year,us.population)
plt.plot(china.year,china.population)
plt.show()
```

```python
plt.plot(us.year,us.population/10**6)
plt.plot(china.year,china.population/10**6)
plt.show()
```

```python
plt.plot(us.year,us.population/10**6)
plt.plot(china.year,china.population/10**6)
plt.legend(['US','CHINA'])
plt.title('US and CHINA population')
plt.xlabel('YEAR')
plt.ylabel('US and CHINA')
plt.show()
```

##  taking first year population as 100% and calculating the population growth

```python
us_first=us.population.iloc[0]
```

```python
us.population / us_first 
```

```python
us.population / us_first * 100
```

```python
china_first=china.population / china.population.iloc[0]
```

```python
china_first * 100
```

```python
plt.plot(us.year,us.population / us_first * 100)
plt.plot(china.year,china_first * 100)
plt.legend(['US','CHINA'])
plt.title('US and CHINA population ')
plt.xlabel('YEAR')
plt.ylabel('US and CHINA (compare from first year population)')
plt.show()
```
